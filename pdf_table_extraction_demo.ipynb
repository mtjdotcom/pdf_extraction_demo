### Page 29 - 2019 Global Startup Ecosystem Ranking

```python
Image(url= "page29.png", width=500, height=400)
```

```python
ranking_page = tabula.read_pdf(pdf_path, pages=29, stream=True,pandas_options={'header': None})
```

```python
r_page_nd = ranking_page.drop_duplicates()
```

```python
r_page_nd.columns = ["City", "Ranking", "Change from 2017", "Performance", "Funding", "Market Reach", "Connectedness", "Talent",
          "Experience", "Knowledge", "Growth Index"]
```

```python
r_page_nd.reset_index(drop=True, inplace=True)
```

```python
r_page_nd = r_page_nd.drop([3,7], axis=0)
```

### Page 54 - Decline Sub-Sectors

```python
Image(url= "page54.png", width=500, height=400)
```

```python
sub_sectors = tabula.read_pdf(pdf_path, pages=54, stream=True)
```

```python
sub_sectors.drop('Decline Sub-Sectors', axis=1, inplace=True)
sub_sectors.drop([0,2,4,6], axis=0, inplace=True)
```

```python
sub_sectors.columns = ["Early Stage Deals 5-Year Growth - Combo", "Exits 5-Year Growth, Count", "Share of Global Startups", "Startup Creation Growth"]
```

```python
sub_sectors['Sub-Sector'] = sub_sectors["Early Stage Deals 5-Year Growth - Combo"].str.split(' ').str[0]
sub_sectors['Early Stage Deals 5-Year Growth'] = sub_sectors["Early Stage Deals 5-Year Growth - Combo"].str.split(' ').str[-1]
```

```python
sub_sectors.drop("Early Stage Deals 5-Year Growth - Combo", axis=1, inplace=True)
```

```python
sub_sectors.iloc[1]['Sub-Sector'] = "Digital Media"
```

```python
sub_sectors = sub_sectors.reindex(columns=['Sub-Sector', 'Early Stage Deals 5-Year Growth', 'Exits 5-Year Growth, Count', 
        'Share of Global Startups', 'Startup Creation Growth'])
```

#### Clean table

```python
sub_sectors
```

### Page 30 - Software Ecosystem Ranking Comparison, 2017-2019
